```python
import pandas as pd 
import numpy as np 

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder
from sklearn.compose import ColumnTransformer 

from imblearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import cross_validate

from xgboost import XGBClassifier

from sklearn.metrics import f1_score, recall_score, precision_score
```

```python
df = pd.read_csv("diabetes_prediction_dataset_cleaned.csv")
```

```python
X = df.drop("diabetes", axis=1)
y = df["diabetes"]
```

```python
num = ["age","bmi","HbA1c_level","blood_glucose_level"]
cat_ohe = ["gender","smoking_history"]
cat_ord = ["age_group","bmi_cat"]
```

```python
sc = StandardScaler()
num_pipeline = Pipeline([("scaler", sc)]) 

ohe = OneHotEncoder(drop= 'first', sparse_output= False)
ohe_pipeline = Pipeline([('OHE Pipeline', ohe)])

age_cat = ["Child", "Teen", "Young Adult", "Middle Aged", "Senior"]
bmi_cat = ["Underweight", "Healthy", "Overweight", "Obese", "Extremely Obese"]

ord = OrdinalEncoder(categories= [age_cat, bmi_cat], dtype=int)
ord_pipeline = Pipeline([('Ordinal Pipeline', ord)])


preprocessing= ColumnTransformer(transformers=[('Num_pipeline',num_pipeline,num),
                                               ('Ohe_Pipeline',ohe_pipeline,cat_ohe),
                                               ('Ord_Pipeline',ord_pipeline,cat_ord),]
                                 ,remainder='passthrough')

preprocessing
```

```python
final_pipeline = Pipeline(steps= [  ('Preprocessing', preprocessing),
                                    ('smote', SMOTE()),
                                    ('XGBoost', XGBClassifier(XGBoost__learning_rate= 0.2,
                                                                XGBoost__max_depth= 4,
                                                                XGBoost__n_estimators= 400,
                                                                XGBoost__subsample= 1.0))])

final_pipeline.fit(X, y)
```

```python
y_proba = final_pipeline.predict_proba(X)[:, 1]

# Sweep thresholds
thresholds = np.arange(0.2, 0.6, 0.01)
results = []
for t in thresholds:
    y_pred = (y_proba >= t).astype(int)
    results.append({
        'threshold': t,
        'recall': recall_score(y, y_pred),
        'f1': f1_score(y, y_pred),
        'precision': precision_score(y, y_pred)
    })

df_results = pd.DataFrame(results)
print(df_results)
```
